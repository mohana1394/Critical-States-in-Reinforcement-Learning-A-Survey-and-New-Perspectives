```python
# Imports:
# --------
import os
import numpy as np
import scienceplots
import matplotlib.pyplot as plt
```

```python
"""
For the DQN we generate reward curves and epsilon decay to show how the model learned
"""

#! DQN training curves:
#! --------------------
dqn_results_path = "../results/DQN/DQN_paper1_results"

# Set plot style to Scientific Plots
plt.style.use(['science', 'ieee'])


# Plot rewards: Generate recent 50 interval average
# -------------
dqn_rewards = np.load(os.path.join(dqn_results_path, "overall_rewards.npy"))

average_reward = []
for idx in range(len(dqn_rewards)):
    avg_list = np.empty(shape=(1,), dtype=int)
    if idx < 50:
        avg_list = dqn_rewards[:idx+1]
    else:
        avg_list = dqn_rewards[idx-49:idx+1]
    average_reward.append(np.average(avg_list))

fig, axis = plt.subplots(1, 1, figsize=(5, 3))
axis.plot(average_reward, color='blue')
axis.set_title("Average reward per episode")
axis.set_xlabel("Episode")
axis.set_ylabel("Average reward")
fig.savefig(os.path.join(dqn_results_path, "paper1_average_rewards.png"), 
            dpi=300, 
            bbox_inches='tight')


# Plot epsilon decay:
# -------------------
dqn_epsilon_decay = np.load(os.path.join(dqn_results_path, "epsilon_decay.npy"))

fig, axis = plt.subplots(1, 1, figsize=(5, 3))
axis.plot(dqn_epsilon_decay, color='blue')
axis.set_title("$\epsilon$-decay")
axis.set_xlabel("Episode")
axis.set_ylabel("$\epsilon$")
fig.savefig(os.path.join(dqn_results_path, "paper1_epsilon_decay.png"), 
            dpi=300, 
            bbox_inches='tight')
```
